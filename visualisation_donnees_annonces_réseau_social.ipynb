```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# prompt: lire le fichier Social_Network_Ads.csv avec pandas
df_social_network = pd.read_csv('/content/Social_Network_Ads.csv')
```

```python
df_social_network
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df_social_network['Age'], kde=True)
plt.title('Distribution des Âges')
plt.xlabel('Âge')
plt.ylabel('Fréquence')
plt.show()
```

```python
# Diagramme à barres pour la répartition des genres
plt.figure(figsize=(10, 6))
sns.countplot(x='Gender', data=df_social_network)
plt.title('Répartition des Genres')
plt.xlabel('Genre')
plt.ylabel('Nombre')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='Gender', y='EstimatedSalary', data=df_social_network)
plt.title('Distribution des Salaires Estimés par Genre')
plt.xlabel('Genre')
plt.ylabel('Salaire Estimé')
plt.show()
```

```python
# Nuage de points pour la relation entre l'âge et le salaire
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='EstimatedSalary', hue='Purchased', data=df_social_network)
plt.title('Relation entre l\'Âge et le Salaire Estimé')
plt.xlabel('Âge')
plt.ylabel('Salaire Estimé')
plt.show()
```

```python
# Diagramme circulaire pour la proportion des achats
plt.figure(figsize=(10, 6))
df_social_network['Purchased'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, colors=['lightblue', 'lightgreen'])
plt.title('Proportion des Achats')
plt.ylabel('')
plt.show()
```
